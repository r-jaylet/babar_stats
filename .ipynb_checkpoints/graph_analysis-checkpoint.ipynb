```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn
from pyvis import network as net
import networkx as nx
```

```python
matrix = pd.read_csv('matrix.csv').rename({'Unnamed: 0': 'name'}, axis=1).set_index('name')
little_matrix = matrix.iloc[1680:1700, 1680:1700]
```

```python
little_matrix
```

```python
edges_list = little_matrix.copy()
edges_list.values[tuple([np.arange(len(edges_list))]*2)] = np.nan
edges_list = edges_list.stack().reset_index()
edges_list = edges_list.rename({'name': 'source',
                                'level_1': 'target',
                                0: 'weight'}, axis=1)
```

```python
edges_list.head(3)
```

```python
net = Network(height="450px", width="100%", bgcolor="#222222", font_color="white", notebook=True)
#net = Network()

# set the physics layout of the network
net.barnes_hut()

sources = edges_list['source']
targets = edges_list['target']
weights = edges_list['weight']

edge_data = zip(sources, targets, weights)
edge_data
```

```python
for e in edge_data:
    src = e[0]
    dst = e[1]
    w = e[2]

    net.add_node(src, src, title=src)
    net.add_node(dst, dst, title=dst)
    net.add_edge(src, dst, value=w)
```

```python
neighbor_map = net.get_adj_list()
```
