### Imports

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sqlite3
import seaborn
seaborn.set()
```

```python
conn = sqlite3.connect('babar_tables')

sql_customer = pd.read_sql_query('''SELECT * FROM babar_server_customer''', conn)
sql_product = pd.read_sql_query('''SELECT * FROM babar_server_product''', conn)
sql_purchase = pd.read_sql_query('''SELECT * FROM babar_server_purchase''', conn)
```

### Format data

```python
df_customer = pd.DataFrame(sql_customer)
df_product = pd.DataFrame(sql_product)
df_purchase = pd.DataFrame(sql_purchase)
```

```python
df = df_purchase[['amount', 'customer_id', 'product_id']].merge(df_customer[['id', 'nickname', 'year']], how='left', left_on='customer_id', right_on='id')
```

```python
df.tail(5)
```

### Short analysis

```python
# analyze those who are linked with bar
#df_customer.groupby('status_id').agg({'id' : 'count'}).sort_values(by=['status_id'])
```

```python
cust = df_customer.groupby('year').agg({'id' : 'count'})
prod = df_product.groupby('price').agg({'id' : 'count'})
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15,6))

cust.plot(kind='bar', ax=ax1)
ax1.set_title('Number of accounts per promo', fontsize = 15)

prod.plot(kind='hist', ax=ax2, bins = 40)
ax2.set_title('Repartition of prices of products', fontsize = 15)

plt.tight_layout()
plt.show()
```

### Rankings

```python
df_ranking_product = df.groupby('product_id').agg({'amount' : 'sum'}).sort_values(by=['amount'])
df_ranking_customer = df.groupby('customer_id').agg({'amount' : 'sum'}).sort_values(by=['amount'])
df_ranking_promo = df.groupby('year').agg({'amount' : 'sum'})
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(15,6))

df_ranking_product[-15:].plot(kind='bar', ax=ax1)
ax1.set_title('Ranking of most money spent for a product', fontsize = 15)
ax1.set_xlabel('product id')

df_ranking_customer[-15:].plot(kind='bar', ax=ax2)
ax2.set_title('Ranking of most money spent by a customer', fontsize = 15)
ax2.set_xlabel('customer id')

df_ranking_promo.plot(kind='bar', ax=ax3)
ax3.set_title('Ranking of most money spent by promo', fontsize = 15)
ax3.set_xlabel('Promotion')

plt.tight_layout()
plt.show()
```

```python
df_ranking_product = df.groupby('product_id').agg({'amount' : 'count'}).sort_values(by=['amount'])
df_ranking_customer = df.groupby('customer_id').agg({'amount' : 'count'}).sort_values(by=['amount'])
df_ranking_promo = df.groupby('year').agg({'amount' : 'count'})
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(15,6))

df_ranking_product[-15:].plot(kind='bar', ax=ax1)
ax1.set_title('Ranking of most product bought', fontsize = 15)
ax1.set_xlabel('product id')

df_ranking_customer[-15:].plot(kind='bar', ax=ax2)
ax2.set_title('Ranking of most product bought by a customer', fontsize = 15)
ax2.set_xlabel('customer id')

df_ranking_promo.plot(kind='bar', ax=ax3)
ax3.set_title('Ranking of most products bought by promo', fontsize = 15)
ax3.set_xlabel('Promotion')

plt.tight_layout()
plt.show()
```

```python
df_beer = df_purchase[['amount', 'product_id']].groupby('product_id').agg({'amount' : 'sum'}).sort_values(by=['amount'], ascending=False).merge(df_product[['id', 'name', 'price']], how='left', left_on='product_id', right_on='id').drop('id', axis=1)
```

```python
df_beer = df_beer[df_beer.amount > 2500]
df_beer = df_beer[~df_beer.name.isin(['Soft', 'Hard', 'Shot', 'Passeport', 'Kro Demi', 'Blanche Demi'])].reset_index().drop('index',axis=1)
df_beer['name'] = df_beer.name.str.replace('Ã©', 'é')
df_beer['name'] = df_beer.name.str.replace('pinte', 'Pinte')
```

```python
df_beer['vol'] = df_beer.apply(lambda x: 0.5 if x['name'][-5:] == 'Pinte' else 0.33, axis=1)
```

```python
df_beer.head(3)
```

```python
degree = [4.2, 8, 5, 5.4, 5, 8.4, 8.5, 5.9, 7, 4, 6.5, 12, 7.2, 5, 5, 
          9, 5.6, 8.4, 8.4, 6, 3.5, 6, 6.5]
df_beer['degree'] = degree
```

```python
df_beer
```

```python
df_beer['ratio_degree'] = round(df_beer.degree / df_beer.vol, 2)
df_beer['ratio_price'] = round(df_beer.price / df_beer.vol, 2)
df_beer['golden_ratio'] = round(df_beer.degree / df_beer.price, 2)
df_beer = df_beer.drop('amount', axis=1)
```
