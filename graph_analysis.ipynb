```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn
from pyvis.network import Network
import networkx as nx
```

```python
matrix = pd.read_csv('matrix.csv').rename({'Unnamed: 0': 'name'}, axis=1).set_index('name')
```

```python
username = '__relouvet__'
threshold = 30
```

```python
tmp = matrix[matrix.index == username].T
user_list = list(tmp[tmp[username] > threshold].index)
user_list = user_list + [username]
matrix = matrix[matrix.index.isin(user_list)].T
sub_matrix = matrix[matrix.index.isin(user_list)]
sub_matrix
```

```python
edges_list = sub_matrix.copy()
edges_list.values[tuple([np.arange(len(edges_list))]*2)] = np.nan
edges_list = edges_list.stack().reset_index()
edges_list
```

```python
edges_list = edges_list.rename({'level_0': 'source',
                                'name': 'target',
                                0: 'weight'}, axis=1)

edges_list = edges_list[edges_list.weight > threshold]
```

```python
net = Network(height="450px", width="100%", bgcolor="#222222", font_color="white", notebook=True)
#net = Network()

# set the physics layout of the network
net.barnes_hut()

sources = edges_list['source']
targets = edges_list['target']
weights = edges_list['weight']

edge_data = zip(sources, targets, weights)
edge_data
```

```python
edge_data
```

```python
for e in edge_data:
    src = e[0]
    dst = e[1]
    w = e[2]
    
    net.add_node(src, src, title=src)
    net.add_node(dst, dst, title=dst)
    net.add_edge(src, dst, value=w)
```

```python
neighbor_map = net.get_adj_list()

# add neighbor data to node hover data
for node in net.nodes:
                node["value"] = len(neighbor_map[node["id"]])
```

```python
net.show_buttons(filter_=['physics'])
```

```python
net.show("first_net.html")
```
